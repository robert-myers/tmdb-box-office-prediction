```python
for i, (x, y) in enumerate(zip(pd.read_csv("data/train.csv").shape, pd.read_csv("data/test.csv").shape)):
    print(x) if x > y else print(y) if i else print(x + y)
```

```python
df = pd.read_csv("data/train.csv").merge(pd.read_csv("data/test.csv"), how="outer")
df.info()
```

```python
# read
train_df = pd.read_csv("data/train.csv")

# clean
train_df.drop("status", axis=1, inplace=True)
train_df = train_df[(train_df["budget"] >= 15000) & (train_df["revenue"] >= 500)]

# calculate
train_df["roi"] = train_df["revenue"] / train_df["budget"]

# describe
train_df.describe(include="all")

# train_df.sort_values(by="revenue").head(10)
```

```python
fig, ax = plt.subplots(figsize=(13, 13), sharex=True, sharey=True)

cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
sns.scatterplot(x="budget", y="revenue",
                     hue="budget", size="revenue",
                     palette=cmap, sizes=(1, 500),
                     data=train_df, ax=ax)

plt.tight_layout()
```
