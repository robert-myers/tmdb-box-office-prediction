```python
from datetime import datetime
from dateutil.relativedelta import relativedelta
import ast
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set(style="darkgrid")


# plt.style.use("seaborn")
# plt.xkcd()

%matplotlib inline
```

```python
df = pd.read_csv("data/train.csv").merge(pd.read_csv("data/test.csv"), how="outer")
df.info()
```

```python
# read
train_df = pd.read_csv("data/train.csv")

# clean
# train_df.drop("status", axis=1, inplace=True)
literal_eval_columns = ["belongs_to_collection", "genres", "production_companies", "production_countries", "spoken_languages", "Keywords", "cast", "crew"]
for column in literal_eval_columns:
    train_df[column] = train_df[column].apply(lambda row: ast.literal_eval(row) if not pd.isna(row) else [])

train_df["release_date"] = pd.to_datetime(train_df["release_date"])
train_df["release_date"] = train_df["release_date"].apply(lambda row: row - relativedelta(years=100) if row >= datetime.now() else row)

train_df = train_df[(train_df["budget"] > 0) & (train_df["revenue"] > 0)]
# train_df = train_df[(train_df["budget"] >= 15000) & (train_df["revenue"] >= 500)]

# calculate
train_df["roi"] = train_df["revenue"] / train_df["budget"]

# describe
train_df.describe(include="all")
```

```python
# train_df.sort_values(by="release_date", ascending=False).head()
train_df.info()
```

```python
fig, ax = plt.subplots(figsize=(16, 16), sharex=True, sharey=True)

cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
sns.scatterplot(x="budget", y="revenue",
                     hue="runtime", size="popularity",
                     palette=cmap, sizes=(1, 500),
                     data=train_df, ax=ax)

plt.tight_layout()
```
